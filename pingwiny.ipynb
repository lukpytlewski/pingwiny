```python
import pandas as pd
penguins = pd.read_csv('penguins.csv')
penguins
```

```python
penguins.info()
```

```python
penguins.dropna(inplace=True)
penguins.info()
```

```python
features = ['CulmenLength','CulmenDepth','FlipperLength','BodyMass']
target = 'Species'
X, y = penguins[features], penguins[target]
print('X:')
display(X)
print('\ny:')
display(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0, stratify=y)
print ('Treningowe obserwacje: %d\nTestowe obserwacje: %d' % (X_train.shape[0], X_test.shape[0]))
```

```python
from sklearn.preprocessing import StandardScaler
scaler_2var = StandardScaler()
X_train_standardized = scaler_2var.fit_transform(X_train)
X_test_standardized = scaler_2var.transform(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
from sklearn.metrics import classification_report
```

```python
metrics_dataframe = pd.DataFrame(columns = ['Model', 'F1_score_test', 'F1_score_train'])
metrics_dataframe
models = []
models_names = []
predictions_proba_list = []
def calculate_metrics(model, name, X_checked, y_checked, X_trained, y_trained):
    models.append(model)
    models_names.append(name)
    global metrics_dataframe
    predictions = model.predict(X_checked)
    predictions_proba = model.predict_proba(X_checked)
    predictions_proba_list.append(predictions_proba[:,1])
    predictions_trained=model.predict(X_trained)

    ############## metryki dla sprawdzanego modelu ################
    # Precision, Recall, F1, Accuracy
    print(classification_report(y_checked, predictions))

    # Confusion matrix
    plt.figure()
    cm = confusion_matrix(y_checked, predictions)
    ax = sns.heatmap(cm, annot=True, cmap='Blues', fmt='.0f')
    ax.set_title('Confusion Matrix\n\n')
    ax.set_xlabel('\nPredicted Values')
    ax.set_ylabel('Actual Values ')
    plt.show()

    f1_metric_test = f1_score(y_checked, predictions, average='micro')
    f1_metric_trained = f1_score(y_trained, predictions_trained, average='micro')
    metrics_dataframe = metrics_dataframe.append({'Model': name, 'F1_score_test': f1_metric_test, 'F1_score_train': f1_metric_trained},
                                                 ignore_index=True)

    
    return metrics_dataframe
```

W przypadku każdego klasyfikatora najpierw optymalizuję określone hiperparametry, a następnie liczę wskaźniki oceniające jakość klasyfikacji dla zestawu najlepszych hiperparametrów.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

model_lr = LogisticRegression(penalty='elasticnet', solver='saga')
params_lr = {'l1_ratio': [0, 0.33, 0.67, 1],
             'C': [0.01, 0.1, 1, 1.5]}
lr_gridsearch = GridSearchCV(model_lr,
                             params_lr,
                             scoring='f1_micro',
                             cv=5,
                             verbose=10, n_jobs=-1)
lr_gridsearch.fit(X_train_standardized, y_train)
print('\nBest hyperparameter:', lr_gridsearch.best_params_)
lr_model_best = lr_gridsearch.best_estimator_

```

```python
calculate_metrics(lr_model_best, 'Logistic Regression', X_test_standardized, y_test, X_train_standardized, y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model_knn = KNeighborsClassifier()

params_knn = {'n_neighbors': [1, 3, 5, 7]}
knn_gridsearch = GridSearchCV(model_knn,
                             params_knn,
                             scoring='f1_micro',
                             cv=5,
                             verbose=10, n_jobs=-1)
knn_gridsearch.fit(X_train_standardized, y_train)
print('\nBest hyperparameter:', knn_gridsearch.best_params_)
knn_model_best = knn_gridsearch.best_estimator_
```

```python
calculate_metrics(knn_model_best, 'KNN', X_test_standardized, y_test, X_train_standardized, y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier

model_tree = DecisionTreeClassifier()

params_tree = {'max_depth': [1, 2, 3, 4], 'min_samples_leaf': [2, 3, 4, 5]}
tree_gridsearch = GridSearchCV(model_tree,
                             params_tree,
                             scoring='f1_micro',
                             cv=5,
                             verbose=10, n_jobs=-1)
tree_gridsearch.fit(X_train, y_train)
print('\nBest hyperparameter:', tree_gridsearch.best_params_)
tree_model_best = tree_gridsearch.best_estimator_
```

```python
calculate_metrics(tree_model_best, 'Tree', X_test, y_test, X_train, y_train)
```

```python
from dtreeviz import trees

viz = trees.dtreeviz(tree_model_best, x_data=X_train, y_data=y_train,
                     target_name='Species', feature_names = X_train.columns)
viz
```

```python
from sklearn.svm import SVC

model_svm = SVC(probability=True)

params_svm = {'C': [0, 0.33, 0.67, 1], 'kernel': ['linear', 'poly', 'rbf', 'sigmoid', 'rbf']}
svm_gridsearch = GridSearchCV(model_svm,
                             params_svm,
                             scoring='f1_micro',
                             cv=5,
                             verbose=10, n_jobs=-1)
svm_gridsearch.fit(X_train_standardized, y_train)
print('\nBest hyperparameter:', svm_gridsearch.best_params_)
svm_model_best = svm_gridsearch.best_estimator_

```

```python
calculate_metrics(svm_model_best, 'SVM', X_test_standardized, y_test, X_train_standardized, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier

model_forest = RandomForestClassifier(n_estimators=1000)

params_forest = {'max_depth': [1, 2, 3, 4], 'min_samples_leaf': [2, 3, 4, 5]}
forest_gridsearch = GridSearchCV(model_forest,
                             params_forest,
                             scoring='f1_micro',
                             cv=5,
                             verbose=10, n_jobs=-1)
forest_gridsearch.fit(X_train, y_train)
print('\nBest hyperparameter:', forest_gridsearch.best_params_)
forest_model_best = forest_gridsearch.best_estimator_
```

```python
calculate_metrics(forest_model_best, 'Random forest', X_test, y_test, X_train, y_train)
```

```python
from sklearn.ensemble import AdaBoostClassifier

model_adaboost = AdaBoostClassifier(base_estimator=SVC(probability=True))

params_adaboost = {'n_estimators': [3, 5, 10, 20]}
adaboost_gridsearch = GridSearchCV(model_adaboost,
                             params_adaboost,
                             scoring='f1_micro',
                             cv=5,
                             verbose=10, n_jobs=-1)
adaboost_gridsearch.fit(X_train_standardized, y_train)
print('\nBest hyperparameter:', adaboost_gridsearch.best_params_)
adaboost_model_best = adaboost_gridsearch.best_estimator_

```

```python
calculate_metrics(adaboost_model_best, 'Adaptive boost', X_test_standardized, y_test, X_train_standardized, y_train)
```

Dla klasyfikatora Voting wybieram trzy modele parametryczne, estymowane na bazie wcześniej określonych hiperparametrów.

```python
from sklearn.ensemble import VotingClassifier

model_voting = VotingClassifier(estimators=[('Logistic Regression', LogisticRegression(penalty='elasticnet', solver='saga', l1_ratio=1, C=1)),
                                            ('SVM', SVC(probability=True, kernel='linear', C=0.67)),
                                            ('AdaBoost', AdaBoostClassifier(base_estimator=SVC(probability=True), n_estimators=3))], voting='soft')
model_voting.fit(X_train_standardized, y_train)
```

```python
calculate_metrics(model_voting, 'Voting classifier', X_test_standardized, y_test, X_train_standardized, y_train)
```
